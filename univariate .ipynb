```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("messy_customer_dataset.xlsx")
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure()
plt.hist(df['Age'], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.barplot(x='Gender', y='AnnualIncome', data=df)
plt.title("Annual Income by Gender")
plt.show()
```

```python
sns.barplot(x='Gender', y='AnnualIncome', data=df)
plt.title("Annual Income by Gender")
plt.show()
```

```python
sns.boxplot(x=df['SpendingScore'])
plt.title("Spending Score Boxplot")
plt.show()
```

```python
sns.histplot(df['AnnualIncome'], kde=True)
plt.title("Income Distribution")
plt.show()
```

```python
df['City'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("City Distribution")
plt.ylabel("")
plt.show()
```
