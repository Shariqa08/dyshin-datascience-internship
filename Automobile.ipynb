```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("automobile_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
price_stats = df.groupby('company')['price'].agg(['max','min'])
price_stats
```

```python
horsepower_stats = df.groupby('company')['horsepower'].agg(['max','min'])
horsepower_stats
```

```python
price_stats['max'].plot(kind='bar')
plt.title("Maximum Car Price by Company")
plt.xlabel("Company")
plt.ylabel("Price")
plt.show()
```
